```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException,ElementClickInterceptedException
from selenium.common.exceptions import StaleElementReferenceException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from itertools import repeat
import warnings
warnings.filterwarnings('ignore')
```

## Yatra.com

```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.maximize_window()
driver.get("https://www.yatra.com/")
# page down
for i in range(0,1):
    actions = ActionChains(driver)
    for _ in range(0,1):
        actions.send_keys(Keys.PAGE_DOWN).perform()
```

```python
places = []
list_ = driver.find_elements_by_xpath("//div[@class='right_data']/div[2]/ul/li")
for i in list_:
    i.click()
    time.sleep(3)
    place_lnk = driver.find_elements_by_xpath("//*[@id='themeSnipe']/section/div[2]/div[2]/div/div/div/div[1]/div/a")
    for i in place_lnk:
        places.append(i.get_attribute('href'))
```

```python
print(len(places))
```

```python
Flight_Name = []
Departure_time = []
Arrival_time = []
Source = []
Destination = []
No_of_Stop = []
Duration = []
Date_of_journey = []
Price = []

for i in places:
    driver.get(i)
    #page down
    for i in range(0,1):
        actions = ActionChains(driver)
        for _ in range(0,1):
            actions.send_keys(Keys.PAGE_DOWN).perform()

    fN = driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span")
    for i in fN:
        Flight_Name.append(i.text)
 
    dt = driver.find_elements_by_xpath("//div[@class='fs-15 bold time']/div")
    for i in dt:
        Departure_time.append(i.text)
        
    at = driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[1]")
    for i in at:
        Arrival_time.append(i.text)
            
    src = driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")
    for i in src:
        Source.append(i.text)
            
    dest = driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")
    for i in dest:
        Destination.append(i.text)
            
    nt = driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span")
    for i in nt:
        No_of_Stop.append(i.text)
            
    durtn = driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']/p")
    for i in durtn:
        Duration.append(i.text)
             
    dtj = driver.find_elements_by_xpath("//div[@class='day-li text-center cursor-pointer pr active font-primary-color']/p[1]")
    for j in dtj:
        Date_of_journey.append(j.text)
        
    price = driver.find_elements_by_xpath("//div[@class='text-right booking-section no-wrap']/div/div/label/div/div[2]")
    for j in price:
        Price.append(j.text)
        
                      
```

```python
yatra = pd.DataFrame({})
yatra['Flight Name'] = Flight_Name
yatra['Departure time'] = Departure_time
yatra['Arrival time'] =  Arrival_time
yatra['Source'] = Source
yatra['Destination'] = Destination
yatra['No of Stop'] = No_of_Stop
yatra['Duration'] =  Duration
yatra['Date of Journey'] = Date_of_journey*10
yatra['Price'] = Price
yatra
```

```python
driver.close()
```

## Easymytrip.com

```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.maximize_window()
driver.get("https://www.easemytrip.com/")
#page down
for i in range(0,1):
    actions = ActionChains(driver)
    for _ in range(0,1):
        actions.send_keys(Keys.PAGE_DOWN).perform()
```

```python
href = []
lnk = driver.find_elements_by_xpath("//div[@class='lft_block']/div/ul/li[1]/a[1]")
lnk2 = driver.find_elements_by_xpath("//div[@class='rit_block']/div/ul/li[1]/a[1]")
for i in lnk:
    href.append(i.get_attribute('href'))
for i in lnk2:
    href.append(i.get_attribute('href'))
    
```

```python
print(len(href))
```

```python
Flight_Name = []
Departure_time = []
Arrival_time = []
Source = []
Destination = []
No_of_Stop = []
Duration = []
Date_of_journey = []
Price = []

for i in href:
    driver.get(i)
    #page down
    for i in range(0,1):
        actions = ActionChains(driver)
        for _ in range(0,1):
            actions.send_keys(Keys.PAGE_DOWN).perform()
            
    try:
        fN = driver.find_elements_by_xpath("//span[@class='txt-r4 ng-binding']")
        for i in fN:
            Flight_Name.append(i.text)
    except NoSuchElementException:
        Flight_Name.append('-')
     
    try:
        dt = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[1]")
        for i in dt:
            Departure_time.append(i.text)
    except NoSuchElementException:
        Departure_time.append('-')
    
    try:
        at = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[1]")
        for i in at:
            Arrival_time.append(i.text)
    except NoSuchElementException:
        Arrival_time.append('-')
        
    try:    
        src = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[2]")
        for i in src:
            Source.append(i.text)
    except NoSuchElementException:
        Source.append('-')
            
    dest = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[2]")
    for i in dest:
        Destination.append(i.text)
    
    try:
        nt = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']/span[2]")
        for i in nt:
            No_of_Stop.append(i.text)
    except NoSuchElementException:
        No_of_Stop.append('-')
    
    try:
        durtn = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']/span[1]")
        for i in durtn:
            Duration.append(i.text)
    except NoSuchElementException:
        Duration.append('-')
        
    try:     
        dtj = driver.find_elements_by_xpath("//div[@class='ai-d fnt17']/label")
        for j in dtj:
            Date_of_journey.append(j.text)
    except NoSuchElementException:
        Date_of_journey.append('-')
    
    try:
        price = driver.find_elements_by_xpath("//div[@class='col-md-12 col-sm-12 main-bo-lis pad-top-bot ng-scope']/div[1]/div[5]/div[1]/div[2]")
        for j in price:
            Price.append(j.text)
    except NoSuchElementException:
        Price.append('-')
        

    #page down
    for i in range(0,1):
        actions = ActionChains(driver)
        for _ in range(0,1):
            actions.send_keys(Keys.PAGE_UP).perform()
```

```python
EMT = pd.DataFrame({})
EMT['Flight Name'] = Flight_Name
EMT['Departure time'] = Departure_time
EMT['Arrival time'] =  Arrival_time
EMT['Source'] = Source
EMT['Destination'] = Destination
EMT['No of Stop'] = No_of_Stop
EMT['Duration'] =  Duration
EMT['Date of Journey'] = Date_of_journey*20
EMT['Price'] = Price
EMT
```

## travolook.in

```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.maximize_window()
driver.get("https://www.travolook.in/")
#page down
for i in range(0,4):
    actions = ActionChains(driver)
    for _ in range(0,1):
        actions.send_keys(Keys.PAGE_DOWN).perform()
```

```python
href = []
lnk = driver.find_elements_by_xpath("//section[@class='hmcontent']/div/ul/li/a")
for i in lnk:
    href.append(i.get_attribute('href'))
```

```python
print(len(href))
```

```python
Flight_Name = []
Departure_time = []
Arrival_time = []
Source = []
Destination = []
No_of_Stop = []
Duration = []
Duration_n = []
Date_of_journey = []
Price = []

for i in href:
    driver.get(i)
    srch = driver.find_element_by_xpath("//*[@id='searchengine_btn']").click()
    # scrape data
    try:
        fN = driver.find_elements_by_xpath("//div[@class='col-3 col-sm-3 pdl0 pdr0 airline']/span")
        for i in fN:
            Flight_Name.append(i.text)            
    except NoSuchElementException:
        Flight_Name.append('-')
     
    try:
        dt = driver.find_elements_by_xpath("//div[@class='col-6 col-sm-6 airtime pdl0 pdr0']/div[1]/b")
        for i in dt:
            Departure_time.append(i.text.split()[0])
    except NoSuchElementException:
        Departure_time.append('-')
    
    try:
        at = driver.find_elements_by_xpath("//div[@class='col-6 col-sm-6 airtime pdl0 pdr0']/div[3]/b")
        for i in at:
            Arrival_time.append(i.text.split()[0])
    except NoSuchElementException:
        Arrival_time.append('-')
        
    try:    
        src = driver.find_elements_by_xpath("//div[@class='col-6 col-sm-6 airtime pdl0 pdr0']/div[1]/b/small")
        for i in src:
            Source.append(i.text)
    except NoSuchElementException:
        Source.append('-')
            
    dest = driver.find_elements_by_xpath("//div[@class='col-6 col-sm-6 airtime pdl0 pdr0']/div[3]/b/small")
    for i in dest:
        Destination.append(i.text)
    
    try:
        nt = driver.find_elements_by_xpath("//div[@class='col-3 timstop text-center pdl0 pdr0']/span/small")
        for i in nt:
            No_of_Stop.append(i.text)
    except NoSuchElementException:
        No_of_Stop.append('-')
    
    try:
        durtn = driver.find_elements_by_xpath("//div[@class='resultwaybx div_outbound']/div/div[1]/div/div[3]/span")
        for i in durtn:
            Duration.append(i.text)
    except NoSuchElementException:
        Duration.append('-')
        
    try:     
        dtj = driver.find_elements_by_xpath("//*[@id='Fly_depdate']")
        for j in dtj:
            Date_of_journey.append(j.get_attribute('value'))
    except NoSuchElementException:
        Date_of_journey.append('-')
    
    try:
        price = driver.find_elements_by_xpath("//div[@class='resultwaybx div_outbound']/div/div[2]/p[1]")
        for j in price:
            Price.append(j.text)
    except NoSuchElementException:
        Price.append('-')    
```

```python
n = []
p = []
F_Name = []
n1 = []
p1 = []
Duration_n = []

for i in Flight_Name:
    n.append(i.split())
for i in n:
    p.append(i.pop())
for i in n:
    F_Name.append(' '.join(i))
    
for i in Duration:
    n1.append(i.split(maxsplit=2))
for j in n1:
    p1.append(j.pop())
    
for k in n1:
    Duration_n.append(' '.join(k))
```

```python
travolook = pd.DataFrame({})
travolook['Flight Name'] = F_Name[0:10000]
travolook['Departure time'] = Departure_time[0:10000]
travolook['Arrival time'] =  Arrival_time[0:10000]
travolook['Source'] = Source[0:10000]
travolook['Destination'] = Destination[0:10000]
travolook['No of Stop'] = No_of_Stop[0:10000]
travolook['Duration'] =  Duration_n[0:10000]
travolook['Date of Journey'] = (Date_of_journey*129)[0:10000]
travolook['Price'] = Price[0:10000]
travolook
```
